```python
import pandas as pd

# Sample dataset
data = {
    'Date': ['2023-09-01', '2023-09-01', '2023-09-02', '2023-09-02', '2023-09-03'],
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Sales': [100, 150, 120, 200, 80]
}

df = pd.DataFrame(data)

# Create a pivot table to summarize data
pivot_table = pd.pivot_table(df, values='Sales', index='Date', columns='Product', aggfunc='sum')

# Display the pivot table
print(pivot_table)
```
